```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
#df_test = pd.read_csv('./data/raw/test.csv',parse_dates=['click_time'], nrows=1000000)
df_test = pd.read_csv('./data/raw/test.csv',parse_dates=['click_time'])
```

```python
df_test.info()
```

```python
df_test.head()
```

```python
df_test['hour'] = df_test['click_time'].dt.hour

df_test = df_test.sort_values(['ip', 'click_time'])
df_test['ip_time_diff'] = df_test['click_time'].diff().dt.seconds
df_test.loc[df_test.ip != df_test.ip.shift(), 'ip_time_diff'] = -1

df_test = df_test.sort_values(['ip', 'app','device','os','channel', 'click_time'])
df_test['all_time_diff'] = df_test['click_time'].diff().dt.seconds
df_test.loc[df_test.channel != df_test.channel.shift(), 'all_time_diff'] = -1
```

```python
df_test.head()
```

```python
df_test['ip_count'] = df_test.groupby('ip')['ip'].transform('count')
df_test['all_count'] = df_test.groupby(['ip', 'app','device','os','channel'])['ip'].transform('count')
```
